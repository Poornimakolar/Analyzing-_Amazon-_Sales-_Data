### Let's create heatmap to find correlation between features.

```python
sns.heatmap(sales_amz.corr(method='pearson'),annot=True, cmap='Blues')
```

As we can see in the above heatmap Unit Price and Unit Cost are Highly corelated.
And Unit Price moderately related to Total Revenue and Total Cost but less related to Total Profit.
lastly their is No relation between Order ID and rest other features.

```python
sales_amz['Item Type'].value_counts()
```

```python
wp = { 'linewidth' : 1, 'edgecolor' : "black" }
labels = ('Cloths','Cosmetics','Office Suppiles','Fruits','Personal Care','Household','Beverages','Baby Food','Cereal','Vegetables','Snacks','Meat')
colors = ('gold','peachpuff','gold','peachpuff','gold','peachpuff','gold','peachpuff','gold','peachpuff','gold','peachpuff')
plt.figure(figsize = (13, 6))
plt.pie(sales_amz['Item Type'].value_counts(),data = sales_amz, 
        autopct = '%0.0f',shadow = False ,
        colors = colors,labels = labels,
        wedgeprops = wp)
plt.axis('equal')
plt.title('Which Item Covers Most Area')
plt.legend(sales_amz['Item Type'].value_counts(), fontsize= 13)
plt.show()
```

```python
colors = {'orange','orchid','orange','orchid'}
wp = {'linewidth': 1, 'edgecolor': 'black'}
sales_amz['Order Priority'].value_counts().plot(kind='pie',colors = colors,wedgeprops = wp, autopct = '%0.0f')
plt.show()
```

```python
sales_amz.groupby('Item Type').mean()['Units Sold'].plot(kind = 'bar',color = 'orchid')
plt.title('Avg. Units sold by Item Type')
plt.xlabel('Item Type')
plt.ylabel('Avg. Units Sold')
plt.show()
```

```python
sales_amz['Order Date'] = pd.to_datetime(sales_amz['Order Date'])
```

```python
sales_amz.groupby('Order Date').sum()['Total Revenue'].plot(kind = 'line',color = 'orchid')
plt.title('Total Revenue over time')
plt.xlabel('Order Date')
plt.ylabel('Total Revenue')
print('*'*20,'Total Revenue:',sales_amz['Total Revenue'].sum(),'*'*23)
plt.show()
```

## Month-wise Sales

```python
sales_amz['Order Month'] = sales_amz['Order Date'].dt.month
```

```python
sales_amz.groupby('Order Month').sum()['Total Revenue'].plot(kind = 'bar', color = 'orchid')
plt.title('Total Revenue By Month')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.show()
```

## Year-wise Sales

```python
sales_amz['Order Year'] = sales_amz['Order Date'].dt.year
```

```python
sales_amz.groupby('Order Year').sum()['Total Revenue'].plot(kind = 'bar', color = 'orchid')
plt.title('Total Revenue By Year')
plt.xlabel('Order Year')
plt.ylabel('Total Revenue')
plt.show()
```

## Relation between Unit Price and Total Revenue

when Unit Price increases Total Revenue Increases.

```python
plt.scatter(sales_amz['Unit Price'],sales_amz['Total Revenue'],color = 'orchid')
plt.title('Unit Price Vs Total Revenue')
plt.xlabel('Unit Price')
plt.ylabel('Total Revenue')
```

```python
sales_amz.groupby('Region')['Total Revenue'].sum().plot(kind = 'bar', color = 'orchid')
plt.title('Total Revenue By Region')
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.show()
```

```python
sns.barplot(x= 'Order Priority', y= 'Total Profit',color = 'orchid', data =sales_amz)
plt.show()
```

## Yearly-Month wise

```python
Yearly_Monthwise = sales_amz.groupby(['Order Month','Order Year']).sum().reset_index()
Yearly_Monthwise.head()
```

```python
sns.catplot(x ='Order Month', y = 'Total Profit', data = Yearly_Monthwise, kind = 'bar',color = 'orchid', col = 'Order Year', col_wrap = 3)
```
